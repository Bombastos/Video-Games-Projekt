## 1. Praxisteil - PreProcessing
Lese den Datensatz ein und analysiere ihn. Dokumentiere deine Schritte. 

```python
# Folgende Bibliotheken werden hier gebraucht
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Einlesen der Datensätze
# Wir haben 2 Datensätze vorgegeben, die wir zusammenführen sollten.
df1 = pd.read_csv('vgsales_teil1.csv', delimiter='|', index_col=0) 
# oben wurden Werte  wo | stand getrennt, damit wir mit Python daraus eine Tabelle erkennen kann
df2 = pd.read_csv('vgsales_teil2.csv',index_col=0)
```

```python
# Zusammenführen von df1 und df2 mit dem concat Befehlt
# Die Concat Funktion eignet sich am besten für unsere Zusammenführung, da die Datenspalten identisch sind.
# Sie verbindet die Daten dann einfach über den Spalten (Spaltenbeschriftung).
# Mann könnte auch die merge Funktion mit Argumenten verwenden.
# Allerdings kann man sich mit dem concat Befehl etwas Tipparbeit ersparen.
# Wäre die Datenausrichtung auf die Indizes (Zeilenbeschriftung) ausgelegt, müsste ich die join Funktion verwenden.
games = pd.concat([df1, df2])
games
```

Der verbundene Datensatz hat also 16598 Zeilen und 10 Spalten

```python
### Auf dem ersten Blick scheinen die Daten nach den Verkaufszahlen sortiert zu sein (siehe Spalte Rank). 
# Diese Spalte werde ich nicht analysieren, daher können wir sie aus unserem DataFrame games löschen.
games.drop("Rank", axis=1,inplace=True)
```

```python
# Fehlende Werte sind mit dem Platzhalter 'Ladenhueter' befüllt. 
# Diese sollten im weiteren Prozess gelöscht bzw. mit NaN ersetzen, um sinnvolle analysen durchzuführen.
games.replace("Ladenhueter", np.nan, inplace=True) 
games.info() # hiermit lassen wir uns die Datentypen aller Spalten anzeigen und die Anzahl der non-null Werte
```

Bei der Spalten Year fehlen 271, bei Publisher 58, bei NA_Sales 4499, bei EU_Sales 5730, bei JP_sales 10455 und bei Other_Sales 6477 Werte.

```python
# Ich will alle Zeilen löschen, wo die Spalte Year Null/NaN Werte hat.
# Die NaN Werte sind entweder wortwörtlich Null-Werte oder Werte, die nicht erhoben wurden.
# Bei den Sales Spalten, nehme ich an, dass es Null-Werte sind, was praktisch auch sinn machen kann.
# Bei den Spalten Year und Publisher kann ich die NaNs allerdings nicht mit Nullen ersetzen, da es kein Sinn macht.
# Da ihr Anteil an den Gesamten Daten gering ist (271 von 16598 bei Year und 58 von 16958 bei Publisher), 
# ist es unwahrscheinlich, dass unser Datensatz dadurch beeinträchtigt wird.
games.dropna(subset=["Year"], axis=0, inplace=True)
games.dropna(subset=["Publisher"], axis=0, inplace=True)


# Wichtig! Index zurücksetzen, da wir Zeilen gelöscht haben
games.reset_index(drop=True, inplace=True)
games.info()
```

Unser Datensatz hat sich damit um 307 weitere Zeilen verkleinert.

```python
# Mit dieser Funktion kann man die Anzahl der Nullwerte in den einzelnen Spalten anzeigen lassen
games.isnull().sum()
```

```python
# Wie würde sich die Dimension verändern, wenn wir alle Zeilen löschen, in denen mindestens 1 Wert fehlt:
games.dropna().shape
# Wie wir im Output unten sehen, würden wir nur noch 2365 statt der ürsprünglichen 16598 Zeilen im DataFrame haben.
# Und das wollen wir verhindern.
```

```python
# Befüllen wir nun die NaNs mit Nullen, um Spalten unten addieren zu können
# und wie oben erklärt könnten wir die NaNs als Nullen interpretieren, da sich Spiele in bestimmten Regionen unbekannt sind
# Im Hinterkopf sollten man aber behalten, dass die NaNs auch dadurch aufkommen, dass das Spiel zwar verkauft wurde,
# aber nicht in unseren Datensatz verzeichnet ist. Das löschen dieser Zeilen würde unseren Datensatz erheblich verkleinern (siehe Zeile drüber) und das wollen wir nicht.
# Die fehlenden Sales Werte mit dem mean oder median zu füllen, wäre ebenfals nicht sinnvoll für die Analyse.
games_neu = games.fillna(0)
games_neu.info()
```

### 2.a) Wie viel % an der Varianz erklären die ersten beiden Komponenten? 
(Wie heißt das Modell, welches du dafür verwenden kannst, führe das Modell aus und interpretiere das Ergebnis)

Das Modell heißt Hauptkomponentenanalyse oder Principal Component Analysis. Sie eignet sich als Dimensionsreduktion und ist eine unsupervised Methode. Sie dient dafür mehrere Variablen, die ähnlich sind in einer Komponente zusammen zufassen. In anderen Worten drückt die PCA aus, welcher Anteil der Varianz der Daten durch die einzelnen Komponenten erklärt wird.

```python
# Erforderliche Bibliotheken importieren
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Um den Standardscaler verwenden zu können, dürfen wir nur numerische Werte in unserer DataFrame haben
# Deswegen berücksichtigen wir hierfür nur Spalten mit floats oder integers
games_num = games_neu[['Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']]
```

```python
# Daten standardisieren, dh auf einer Ebene bringen mit standardScaler
scaler = StandardScaler()

# unsere Daten sind in games_neu gespeichert, daher können wir hier games_neu verwenden

games_num_scaled = scaler.fit_transform(games_num)
```

```python
# PCA durchführen, um mit weniger Features weiterrechnen zu können

# wir lassen uns 4 Hauptkomponenten anzeigen

pca4 = PCA(n_components = 4)

pca4.fit(games_num_scaled)   # games_neu beinhaltet mittlerweile unsere standardisierten Features

# wir schauen uns vor der Transformation ersteinmal an mit wie vielen Komponenten wir weiter gehen

pca4_variance = pca4.explained_variance_ratio_
print(pca4_variance)
```

#### Zwischenfazit: 
die erste (61%) und zweite Hauptkomponente (18%) erklären zusammen 79.43% an der Varianz.

```python
# gibt aus, mit welchem Faktor die Varianz der jeweiligen Features in die jeweiligen Komponenten eingehen

pd.DataFrame(pca4.components_, columns=['Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales'], index=['1','2','3','4'])
```

### Fazit: 
Es wäre ratsam auch die 3te Hauptkomponente zu berücksichtigen, da sie 11% der Varianz erklärt und sie zusammen mit den ersten beiden Hauptkomponenten über 90% der Varianz erklären würde. Wie oben zu sehen ist, werden die Verkäufe in Japan stark von der 3ten Hauptkomponente erklärt. Im allgemeinen kommt es auf die Präferenz des Analysten an, wie viele Prozent der Varianz sein modell erklären soll.

### 2.b) Wie viele Cluster existieren? 
(Wie heißt das Modell? Ordne das Modell thematisch ein. Wie kannst du herausfinden, ob dein Modell "gut" ist?)
Experimentiere mind. an 3 Clustern.
Visualisiere dein Ergebnis mit der Heatmap. Wie ist die Heatmap zu interpretieren? Was sagt sie aus?
Hinweis: Du kannst eine Pipeline verwenden.

Das Modell zum Aufspühren von Clustern heißt K-Means. Sie ist eine unsupervised machine learning Methode, mit der man feststellen kann wie viele untergruppen angemessen sind. Wichtig ist zu wissen, dass der Algorithmus der hinter K-Means steckt nicht-deterministisch ist. D.h. es kann 2 unterschiedliche Ergebnisse bei 2 getrennten Durchgängen liefern, selbst wenn die Durchgänge auf den selben Input basieren. Um die Performance des Modells zu messen, verwenden wir die Confusion Matrix. Je höher die Werte an der Diagonale, desto besser ist unsere Modell. Umgekehrt ist unser Modell weniger gut, wenn die Werte an der nicht-Diagonale der Matrix hoch sind.
Die Confusion matrix können wir anhand der Heatmap visualisieren. Anhand einer Farbscala, kennzeichnet die Heatmap die unterschiedlichen Werte, sodass man farblich die Güte des Clustering Methode erkennt.

```python
# Das sind zusätliche Bibliotheken, die wir hinzufügen wollen, um Kmeans z
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix


from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
# K-Means arbeitet mit numerischen Daten, deswegen berücksigten wir unsere Sales Spalten, welche die Anzahl der verkauften Spiele Kopien widergibt.
games_kmean= games_neu.iloc[:,[5,6,7,8,9]]

# Wir brauchen wieder den standardscaler womit wir die Daten auf einer Ebene bringen  
scaler = StandardScaler()
games_scaled = scaler.fit_transform(games_kmean)

# Außerdem geben wir 3 Cluster mit an
n = 3
cluster = KMeans(n_clusters=n,n_init = 100, max_iter=500)
# fitten der kmeans cluster mit den skalierten games daten 
cluster.fit(games_scaled)
# Spalte mit cluster zufügen
games_kmean['Cluster']=cluster.predict(games_scaled)

games_kmean.head()
```

```python
# Verteilung ansehen
sns.pairplot(games_kmean, hue='Cluster')
```

Hier komme ich nicht weiter, da ich keine Kategorischen Daten habe und somit die Confusion matrix nicht aufstellen kann. Aber man kann schon sehen, dass er die Nullwerte in einem Cluster zusammengepackt hat (blauer Cluster).Es macht auch kein Sinn mit der Cut Funktion die Genres zu kategorisieren, da Python unmöglich anhand der strings die Genres sinnvolle Kategorien bilden kann. Stattdessen verwende ich ein anderes Anwendungsmodell (Pipeline) an. Deren Performance messe ich dann mit dem silhouette_score und dem ARI Score.

```python
# Um die Genres in den Auswertemethoden zu verwenden,
# müssen Genre strings mit LabelEncoder in ganze Zahlen umgewandelt werden
true_genres= games_neu.iloc[:,3]
label_encoder = LabelEncoder()

# Fitten des label encoders an die Genre Spalte nicht vergessen.
true_labels = label_encoder.fit_transform(true_genres)

true_labels[45:65]
```

```python
# Mit dem Suffix classes_ können wir uns die Klassennamen anschauen
label_encoder.classes_
```

```python
# die Länge der clustern anhand der länge der Klassen bestimmen. 
n_clusters = len(label_encoder.classes_)
```

```python
# wir bauen 2 Pipelines. 1 für Scaler und PCA und dadrunter eine pipeline für KMeans
# Da wir keine normalverteilung annehmen nutzen wir den MinMaxScaler statt StandardScaler.
preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=2, random_state=42)),
    ]
)
```

```python
clusterer = Pipeline(
   [
       (
           "kmeans",
           KMeans(
               n_clusters=12,
               init="k-means++",
               n_init=50,
               max_iter=500,
               random_state=42,
           ),
       ),
   ]
)
```

```python
# zusammenpacken der pipelines
pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)
```

```python
# fiten der pipelines an unsere numerischen Sales spalten
pipe.fit(games_kmean)
```

```python
# Die Performance des clusterings messen wir anhand des Silhouettenkoeffizienten:
preprocessed_data = pipe["preprocessor"].transform(games_kmean)

predicted_labels = pipe["clusterer"]["kmeans"].labels_

silhouette_score(preprocessed_data, predicted_labels)
```

```python
# Da die Groundtruth Genres vorhanden sind, können wir auch den ARI berechnen:
adjusted_rand_score(true_labels, predicted_labels)
```

```python
mat = confusion_matrix(true_labels, predicted_labels)
games_kmean_mat = pd.DataFrame(mat, columns=[label_encoder.classes_],index=[label_encoder.classes_])
games_kmean_mat
```

```python
plt.figure(figsize=(12, 14))
sns.set(font_scale=1.2)
plt.subplot(211)
sns.heatmap(mat, annot=True,xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_, fmt='d')
plt.title("Confusion Matrix Heatmap")

plt.show()
```

#### Fazit:
Beide Gütemaße gehen von -1 bis 1. Ein Silhouettenkoeffizient von 0 zeigt an, dass sich die Cluster signifikant überlappen, und ein Silhouettenkoeffizient von 1 zeigt an, dass die Cluster gut voneinander getrennt sind. In unserem Beispiel mit 12 Clustern sind die Genres teilweise von einander zu unterscheiden (Silhouettenkoeffizient=0.674). Ein ARI-Score von 0 bedeutet, dass Cluster-Genres zufällig vergeben werden, und ein ARI-Score von 1 bedeutet, dass die wahren und die vorhergesagten Genres identische Cluster bilden. In unserem Beispiel werden die Genres Cluster zufällig vergeben (ARI=0.005 bei 12 Clustern). Durch das ändern der Cluster anzahl, habe ich festgestellt, dass der Silhouettenkoeffizient mit weniger Clustern tendenziell steigt und der ARI Wert mit weniger Clustern abnimmt.

```python
# Mit crosstab wird eine einfache Kreuztabellierung von zwei (oder mehr) Faktoren berechnet. 
# Standardmäßig berechnet sie eine Häufigkeitstabelle der Features.
Genre_Platf = pd.crosstab(games_neu.Platform,games_neu.Genre) # Wir füttern sie mit Genres und Platformen
Genre_PlatfTotal = Genre_Platf.sum(axis=1).sort_values(ascending = False)# Summe bilden und dann absteigend sortieren

plt.figure(figsize=(10,7))
sns.set_style("darkgrid", {"axes.facecolor": ".9"})


ax=sns.barplot(y=Genre_PlatfTotal.index, x=Genre_PlatfTotal.values, orient='h')
ax.set(title='Anzahl veröffentlichter Spiele pro Platform',xlabel='Anzahl Spiele', ylabel='Platform')


plt.show()
```

### Analyse:
Anhand dieser Graphik können wir erkennen, dass Nintendo DS und die PS2 die meisten Videospiele herrausgebracht haben. Beide haben jeweils über 2000 Spiele veröffentlicht, während die anderen Platformen höchstens 1300 Spiele rausgebracht haben.
